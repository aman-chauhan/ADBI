# Degree Distribution Plots

## Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
%matplotlib inline
```

## Baseline Power Law with lambda=2 and lambda=3 and constant 1

```python
x = [i for i in range(1,50)]
y2 = [1/math.pow(i, 2) for i in range(1,50)]
y3 = [1/math.pow(i, 3) for i in range(1,50)]

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f.set_figheight(6)
f.set_figwidth(16)
f.suptitle('Baseline Power Law Curves')
ax1.set(xlabel='i.i.d', ylabel='count')
ax1.set_title('Power Law with lambda=2')
ax1.plot(x, y2)
ax2.set(xlabel='i.i.d', ylabel='count')
ax2.set_title('Power Law with lambda=3')
ax2.plot(x, y3)
plt.show()
```

## Random Plots

```python
types = {'degree': int, 'count': int}
gnp1 = pd.read_csv('gnp1.csv', names=['i.i.d', 'degree', 'count']).drop('i.i.d', 1).drop(0,0).astype(types)
gnp2 = pd.read_csv('gnp2.csv', names=['i.i.d', 'degree', 'count']).drop('i.i.d', 1).drop(0,0).astype(types)
gnm1 = pd.read_csv('gnm1.csv', names=['i.i.d', 'degree', 'count']).drop('i.i.d', 1).drop(0,0).astype(types)
gnm2 = pd.read_csv('gnm2.csv', names=['i.i.d', 'degree', 'count']).drop('i.i.d', 1).drop(0,0).astype(types)
```

```python
f, ax = plt.subplots(2, 2)
f.set_figheight(12)
f.set_figwidth(16)
ax[0,0].set_title('gnp1')
ax[0,0].set(xlabel='degree', ylabel='count')
ax[0,0].plot(gnp1['degree'], gnp1['count']/sum(gnp1['count']))
ax[0,1].set_title('gnp2')
ax[0,1].set(xlabel='degree', ylabel='count')
ax[0,1].plot(gnp2['degree'], gnp2['count']/sum(gnp2['count']))
ax[1,0].set_title('gnm1')
ax[1,0].set(xlabel='degree', ylabel='count')
ax[1,0].plot(gnm1['degree'], gnm1['count']/sum(gnm1['count']))
ax[1,1].set_title('gnm2')
ax[1,1].set(xlabel='degree', ylabel='count')
ax[1,1].plot(gnm2['degree'], gnm2['count']/sum(gnm2['count']))
plt.show()
```
